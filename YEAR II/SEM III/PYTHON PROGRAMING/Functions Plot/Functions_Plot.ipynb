```
'''
13. Write a program that makes use of a function to display sine, cosine,
polynomial and exponential curves.
'''
import math
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
def sin_graph():
  plt.subplot(2, 2, 1)
  plt.plot([math.sin(math.radians(i)) for i in range(0, 361, 1)])
  plt.title("SINE GRAPH")
  plt.ylabel("sin(x)")
  plt.xlabel("value of x")
  plt.grid()
```

```
def cos_graph():
  plt.subplot(2, 2, 2)
  plt.plot([math.cos(math.radians(i)) for i in range(0, 361, 1)])
  plt.title("COSINE GRAPH")
  plt.ylabel("cos(x)")
  plt.xlabel("value of x")
  plt.grid()
```

```
def exp_graph():
  plt.subplot(2, 2, 3)
  plt.plot(np.array([math.exp(x) for x in np.arange(0, 11, 0.4)]))
  plt.title("EXPONENTIAL GRAPH")
  plt.xlabel("value of x")
  plt.ylabel("exp(x)")
  plt.xlim(0, 30)
  plt.grid()
```

```
def poly_graph():
  plt.subplot(2, 2, 4)
  plt.plot([5*i**2+3*i-7 for i in range(1, 100, 1)])
  plt.title("POLYNOMIAL GRAPH")
  plt.xlabel("value of x")
  plt.ylabel("5x^2 + 3x - 7")
  plt.xlim(0, 100)
  plt.grid()
```

```
def main():
  sin_graph()
  cos_graph()
  exp_graph()
  poly_graph()
  plt.tight_layout()
  plt.show()

if __name__ == '__main__':
  main()
```
